```python
def bar(data):
    score_counts = data['writing score'].value_counts().sort_index()

    plt.figure(figsize=(30, 15))
    score_counts.plot(kind='bar', color='brown', width=0.5)
    plt.title('Distribution of Writing Scores', fontsize=30)
    plt.xlabel('Writing Score', fontsize=30)
    plt.ylabel('Number of Students', fontsize=30)
    plt.xticks(rotation=0, fontsize=13)
    plt.yticks(rotation=0, fontsize=13)

   
```

```python
bar(df)
```

```python
def line_graph(data):
    avg_writing_scores = data.groupby('parental level of education')['writing score'].mean().sort_values()
    
    plt.figure(figsize=(12, 6))
    avg_writing_scores.plot(kind='line', marker='o', color='brown')
    plt.title('Average Writing Scores Based on Parents', fontsize=17)
    plt.xlabel('Parents Educational Qualifications', fontsize=17)
    plt.ylabel('Average Writing Score', fontsize=17)
    plt.grid()
   

```

```python
line_graph(df)
```

```python
def heatmap(data):
    corr = data[['math score', 'reading score', 'writing score']].corr()
    plt.figure(figsize=(12, 12))
    heatmap = plt.matshow(corr, cmap='Blues')
    plt.colorbar(heatmap)
    plt.xticks(np.arange(len(corr.columns)), corr.columns, rotation=0)
    plt.yticks(np.arange(len(corr.columns)), corr.columns)
    plt.xticks(rotation=0, fontsize=8)
    plt.title('Correlation Heatmap', fontsize=16)
  
```

```python
heatmap(df) 
```
